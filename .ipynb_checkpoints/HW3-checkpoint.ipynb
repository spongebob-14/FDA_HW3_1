function def

for str to mon(use by train and test)

for count day(use while fill the blank of date)

```python
def str_to_mon(str):
    return {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12',
    }.get(str,'error')

def getday(y=2017,m=8,d=15,n=0):
    the_date = datetime.datetime(y,m,d)
    result_date = the_date + datetime.timedelta(days=n)
    d = result_date.strftime('%Y-%m-%d')
    return d
```

# Logistic Regression

reference:https://medium.com/@jacky308082/machine-learning-%E4%B8%8B%E7%9A%84-logistic-regression-%E5%AF%A6%E4%BD%9C-%E4%BD%BF%E7%94%A8python-d19b971ff9dc

```python
from sklearn.model_selection import KFold 
from sklearn.metrics import accuracy_score

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = [] 

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    clf=LogisticRegression(fit_intercept=True,C=1e15)
    clf.fit(train_x_split,train_y_split)
    #print(clf.intercept_,clf.coef_)
    
    train_pred_y = clf.predict(train_x_split)       
    train_acc = accuracy_score(train_y_split,         
                               train_pred_y)
    valid_pred_y = clf.predict(valid_x_split)       
    valid_acc = accuracy_score(valid_y_split,         
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)


    
    
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
```

```python
pred_y=clf.predict(test_x)

acc = accuracy_score(test_y,pred_y)

print(acc)
```

# Neural Network

reference:https://machine-learning-python.kspax.io/neural_network_models_supervised

```python
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score


train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = [] 

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)

for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                     hidden_layer_sizes=(5000,500), random_state=1)

    clf.fit(train_x_split,train_y_split)
    #print(clf.intercept_,clf.coef_)
    
    train_pred_y = clf.predict(train_x_split)       
    train_acc = accuracy_score(train_y_split,         
                               train_pred_y)
    valid_pred_y = clf.predict(valid_x_split)       
    valid_acc = accuracy_score(valid_y_split,         
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)


    
    
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
```

```python
clf.fit(train_x,train_y)
pred_y=clf.predict(test_x)

acc = accuracy_score(test_y,pred_y)

print(acc)
```

# XGBoost

會選這個模型只是因為之前聽說他很強，想說試試看XD

boosting disicion tree就是對上次做錯的部分提高權重再繼續訓練
而根據我查到的資料，XGBoost特別的地方在於

他會對誤差值再進行學習，這樣加到最後誤差應該會超小

還有就是他再找分裂節點的演算法也很適合(分析分裂之後有沒有更好，也有設閾值避免一直分裂)

Shrinkage 在每次疊代時縮減每個子節點的權重，留空間給後面生成的樹

Column Subsampling 不是很確定我有沒有理解錯，看起來像是每次再取feature的時候隨機選，所以訓練出來的子樹就會學到各種特徵，不會overfitting在某幾種特徵上面

reference:

https://www.itread01.com/content/1545533828.html

https://www.itread01.com/articles/1476146171.html

https://kknews.cc/zh-tw/news/grejk5m.html

```python
from xgboost.sklearn import XGBClassifier

train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = [] 

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)

clf = XGBClassifier(
        #樹的個數
        n_estimators=100,
        # 學習率
        learning_rate= 0.3, 
        # 構建樹的深度，越大越容易過擬合    
        max_depth=6, 
        # 隨機取樣訓練樣本 訓練例項的子取樣比
        subsample=1, 
        # 用於控制是否後剪枝的引數,越大越保守，一般0.1、0.2這樣子
        gamma=0, 
        # 控制模型複雜度的權重值的L2正則化項引數，引數越大，模型越不容易過擬合。
        reg_lambda=1,  
        
        #最大增量步長，我們允許每個樹的權重估計。
        max_delta_step=0,
        # 生成樹時進行的列取樣 
        colsample_bytree=1, 

        # 這個引數預設是 1，是每個葉子裡面 h 的和至少是多少，對正負樣本不均衡時的 0-1 分類而言
        # 假設 h 在 0.01 附近，min_child_weight 為 1 意味著葉子節點中最少需要包含 100 個樣本。
        #這個引數非常影響結果，控制葉子節點中二階導的和的最小值，該引數值越小，越容易 overfitting。
        min_child_weight=1, 

        #隨機種子
        seed=1000 
)



for train_index, valid_index in kf.split(train_x):    # 每個迴圈都會產生不同部份的資料
    train_x_split = train_x.iloc[train_index]         # 產生訓練資料
    train_y_split = train_y.iloc[train_index]         # 產生訓練資料標籤
    valid_x_split = train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    #clf is set outside
    clf.fit(train_x_split,train_y_split)
    #print(clf.intercept_,clf.coef_)
    
    train_pred_y = clf.predict(train_x_split)       
    train_acc = accuracy_score(train_y_split,         
                               train_pred_y)
    valid_pred_y = clf.predict(valid_x_split)       
    valid_acc = accuracy_score(valid_y_split,         
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)


    
    
print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度
    np.min(train_acc_list),                           # 輸出最低訓練準確度
    np.max(train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)     https://www.itread01.com/articles/1476146171.html                       # 輸出最高驗證準確度
))
```
